Este codigo le um arquivo.txt e separa em 3 array com as temperatura media, minima e maxima, e ent√£o cria um grafico cartesiano e de barra para cada cidade.

```python
import matplotlib.pyplot as grafico
import numpy as np

cont=0
listaMes = np.arange(1,13)
MlistaMedia = []
MlistaMinimo = []
MlistaMaixmo = []
RJlistaMedia = []
RJlistaMinimo = []
RJlistaMaximo = []
with open("dadosClimaticos Maca√©.txt","r")as arquivoMacae:
  dadosMacae = arquivoMacae.read()
  dadosMacae = dadosMacae.split()

# print(dadosMacae)
# print('\n')

# Preenche um arry com os dados climaticos de Maca√©
for dadosMacae in dadosMacae:
  if cont > 13 and cont < 26:
    MlistaMedia.append(dadosMacae)
  if cont > 26 and cont < 39:
    MlistaMinimo.append(dadosMacae)
  if cont > 39:
    MlistaMaixmo.append(dadosMacae)

  cont += 1

# print(MlistaMedia)
cont =0

with open("dadosClimaticosRJ.txt","r")as arquivoRJ:
  dadosRJ = arquivoRJ.read()
  dadosRJ = dadosRJ.split()

# print(dadosRJ)
# print('\n')

# Preenche um arry com os dados climaticos do RJ
for dadosRJ in dadosRJ:
  if cont > 13 and cont < 26:
    RJlistaMedia.append(dadosRJ)
  if cont > 26 and cont < 39:
    RJlistaMinimo.append(dadosRJ)
  if cont > 39:
    RJlistaMaximo.append(dadosRJ)
  cont += 1

################################################################################
# Converte os valores do array de string em float
MlistaMediaF = []
MlistaMediaF = list(map(float,MlistaMedia))
MlistaMinimoF = []
MlistaMinimoF = list(map(float,MlistaMinimo))
MlistaMaximoF = []
MlistaMaximoF = list(map(float,MlistaMaixmo))

RJlistaMediaF = []
RJlistaMediaF = list(map(float,RJlistaMedia))
RJlistaMinimoF = []
RJlistaMinimoF = list(map(float,RJlistaMinimo))
RJlistaMaximoF = []
RJlistaMaximoF = list(map(float,RJlistaMaximo))

################################################################################
# Cria√ß√£o do grafico cartesiano
################################################################################
grafico.figure(figsize=(12,8))
grafico.ylabel("T/¬∫C")
grafico.xlabel("Mes")
grafico.grid()
grafico.title("Temperatura de Maca√©")
grafico.plot(listaMes,
             MlistaMediaF, listaMes, MlistaMinimoF, listaMes ,MlistaMaximoF)
grafico.legend(['TMedia','TMin','TMax'])
grafico.savefig("Grafico Macae.png")
################################################################################
grafico.figure(figsize=(12,8))
grafico.ylabel("T/¬∫C")
grafico.xlabel("Mes")
grafico.grid()
grafico.title("Temperatura de RJ")
grafico.plot(listaMes,
             RJlistaMediaF, listaMes, RJlistaMinimoF, listaMes ,RJlistaMaximoF)
grafico.legend(['TMedia','TMin','TMax'])
grafico.savefig("Grafico RJ.png")
```

```python
#B)Cria√ß√£o do grafico em Barra
################################################################################
import matplotlib.pyplot as graficoBarra

width = 0.3

graficoBarra.figure(figsize=(12,8))

grafico.title("Temperatura de Maca√©")
graficoBarra.bar(listaMes-0.3, MlistaMediaF, width, color='orange')
graficoBarra.bar(listaMes, MlistaMinimoF, width, color='cyan')
graficoBarra.bar(listaMes+0.3, MlistaMaximoF, width, color='red')
graficoBarra.xlabel('Mes')
graficoBarra.ylabel('temp C¬∫')
grafico.legend(['TMedia','TMin','TMax'])
grafico.savefig("Grafico barra Macae.png")
graficoBarra.show()


################################################################################
graficoBarra.figure(figsize=(12,8))

grafico.title("Temperatura de RJ")
graficoBarra.bar(listaMes-0.3, RJlistaMediaF, width, color='orange')
graficoBarra.bar(listaMes, RJlistaMinimoF, width, color='cyan')
graficoBarra.bar(listaMes+0.3, RJlistaMaximoF, width, color='red')
graficoBarra.xlabel('Mes')
graficoBarra.ylabel('temp C¬∫')
grafico.legend(['TMedia','TMin','TMax'])
grafico.savefig("Grafico barra RJ.png")
graficoBarra.show()
```
